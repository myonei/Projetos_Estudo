# Airline passenger satisfaction dataset

### Conteúdo deste dataset

| Feature | Detalhes |
| -: | :- |
|**Gender**| Sexo do passageiro (Female, Male)|
|**Customer Type**| Tipo de cliente (Loyal customer, disloyal customer)
|**Age**| A idade do passageiro
|**Type of Travel**| Propósito da viagem (Personal Travel, Business Travel)
|**Class**| Cabine referente à passagem comprada (Business, Eco, Eco Plus)
|**Flight distance**| Distância percorrida na viagem
|**Inflight wifi service**| Nível de satisfação com o wifi (0**\|Not Applicable;1-5)
|**Departure/Arrival time convenient**| Nível de satisfação com relação ao horário de saída e chegada
|**Ease of Online booking**| nível de satisfação com as reservas online
|**Gate location**| Nível de satisfação com a localização do portão
|**Food and drink**| Nível de satisfação com comidas e bebidas
|**Online boarding**| Nível de satisfação com online boarding
|**Seat comfort**| Nível de satisfação com o conforto dos assentos
|**Inflight entertainment**| Nível de satisfação com o entretenimento durante o voo
|**On-board service**| Nível de satisfação com o serviço on-board
|**Leg room service**| Nível de satisfação com relação ao espaço para as pernas
|**Baggage handling**| Nível de satisfação com o cuidado com as bagagens
|**Check-in service**| Nível de satisfação com os serviços de check in
|**Inflight service**| Nível de satisfação durante o voo
|**Cleanliness**| Nível de satisfação com a limpeza
|**Departure Delay in Minutes**| Atrasos para saída do voo em minutos
|**Arrival Delay in Minutes**| Atrasos na chegada em minutos
|**Satisfaction**| Nível de satisfação(Satisfaction, neutral or dissatisfaction)



### Objetivo

Definir quais são os principais fatores que estão mais correlacionados com a satisfação de um passageiro e elaborar um modelo que consegue prever a satisfação de um passageiro.

Para definir o nível de satisfação dos passageiros, utilizaremos as seguintes classes:

   **0**: Neutro ou não satisfeito <br>
   **1**: Satisfeito

```python
train = X_train
train['satisfaction'] = y_train
train
```

```python
# Cálculo do intervalo interquartil
Q1 = train.quantile(0.25)
Q3 = train.quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
import warnings
warnings.filterwarnings("ignore")
#Removendo outliers
train_without_outliers = train[~((train < (Q1 - 1.5 * IQR)) |(train > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
train_without_outliers.shape
```

Com a remoção de outliers, perdemos 25.420 linhas de informações no geral. Neste sentido, em vez de remover todos os outliers, podemos remover apenas de algumas colunas onde há uma grande amplitude.

Conforme abaixo, as features com maior amplitude são:

Flight Distance, Departure Delay in Minutes, Arrival Delay in Minutes

```python
df.describe()
```

```python
remove_outliers = ['Flight Distance','Departure Delay in Minutes', 'Arrival Delay in Minutes']

for col in train:
    if col in remove_outliers:
        q1 = train[col].quantile(0.25)
        q3 = train[col].quantile(0.75)
        IQR = q3 - q1
        
        lower_limit = q1 - 1.5 * IQR
        upper_limit = q3 + 1.5 * IQR
        train = train[train[col] >= lower_limit]
        train = train[train[col] <= upper_limit]
        
```

```python
train.shape
```

Conforme acima, houve mais remoções, mas conseguimos reter um pouco mais de informação.

# Modelos

Os processos a seguir estão divididos nas seguintes etapas:
    1 - Pré-processamento dos dados em uma pipeline
    2 - Treinar um modelo de regressão logística como baseline e verificar as métricas
    3 - Treinar modelos de Machine Learning classicos e verificar o desempenho de cada um
    4 - Treino de um modelo de rede neural.
    

```python
#Separar os nomes das features numéricas e categóricas
num_feat = list(df_num.columns)
cat_feat = list(col_categ)
```

```python
#Instanciando um pre processador que preenche nulos com valores da mediana, uma vez que verificamos a presença de outliers
#Normaliza com standard scaler, as variáveis numéricas e transforma as variáveis categóricas em dummies.
num_transformer = Pipeline(
    steps=[
        ('imputer', SimpleImputer(strategy="median")),
        ('scaler', StandardScaler())
    ]
)

cat_transformer = OneHotEncoder(handle_unknown='ignore')

preprocessor = ColumnTransformer(
    transformers=[
        ("num", num_transformer, num_feat),
        ("cat", cat_transformer, cat_feat),
    ]
)
```

```python
#Treinar o nosso modelo de regressão logística, que servirá como baseline
pipe_logit = Pipeline([("preprocessor", preprocessor),
                       ("logit", LogisticRegression())])

pipe_logit.fit(X_train, y_train)
```

```python
def clf_metrics(modelo, X, y_true, label):
    '''
    Função para mostrar as métricas de avaliação de um modelo pelo classification report
    '''
    print(f"\nMétricas de avaliação de {label}:\n")    
    y_pred = modelo.predict(X)
    print(classification_report(y_true, y_pred))
```

```python
clf_metrics(pipe_logit, X_train, y_train, "treino")
print("#"*80)
clf_metrics(pipe_logit, X_test, y_test, "teste")
```

```python
def clf_metrics_com_return(modelo, X, y_true, label):
  
   
    y_pred = modelo.predict(X)
    
    return classification_report(y_true, y_pred, output_dict=True)
```

```python
def experimentos_ciclo_de_modelagem(dict_pipes,
                                    print_progress = False):
    
    resultado_experimentos = {"estimador" : [],
                              "f1_treino" : [],
                              "f1_teste" : []}

    for label, pipe in dict_pipes.items():

        if print_progress:
            print("\n")
            print("="*80)
            print(f"Estimador: {label}".center(80))
            print("(com hiperparâmetros default)".center(80))
            print("="*80)
            print("\n")

        # ================================

        pipe.fit(X_train, y_train)

        # ================================
        # Passo 2

        dict_metricas_treino = clf_metrics_com_return(pipe, X_train, y_train, "treino")

        if print_progress:
            print("#"*80)

        dict_metricas_teste = clf_metrics_com_return(pipe, X_test, y_test, "teste")

        # pegar as métricas
        f1_treino = dict_metricas_treino["weighted avg"]["f1-score"]
        f1_teste = dict_metricas_teste["weighted avg"]["f1-score"]

        # ================================
        # guardando os resultados do experimento

        resultado_experimentos["estimador"].append(label)
        resultado_experimentos["f1_treino"].append(f1_treino)
        resultado_experimentos["f1_teste"].append(f1_teste)


    df_results = pd.DataFrame(resultado_experimentos)

    df_results["gap"] = (df_results["f1_treino"] - df_results["f1_teste"]).apply(lambda x: x if x > 0 else np.inf)

    df_results = df_results.sort_values("f1_teste", ascending=False).sort_values("gap")

    return df_results
```

```python
pipe_logit = Pipeline([("preprocessor", preprocessor),
                       ("logit", LogisticRegression(random_state=42))])

pipe_rf = Pipeline([("preprocessor", preprocessor),
                    ("rf", RandomForestClassifier(random_state=42))])

pipe_dt = Pipeline([("preprocessor", preprocessor),
                    ("dt", DecisionTreeClassifier(random_state=42))])

pipe_svm = Pipeline([("preprocessor", preprocessor),
                     ("svm", SVC(random_state=42))])

pipe_xbgoost = Pipeline([("preprocessor", preprocessor),
                         ("xgboost", XGBClassifier(random_state=42))])

pipe_lgbm = Pipeline([("preprocessor", preprocessor),
                      ("lgbm", LGBMClassifier(random_state=42))])

pipe_knn = Pipeline([("preprocessor", preprocessor),
                     ("knn", KNeighborsClassifier())])

# =======================================

dict_pipes = {"logit" : pipe_logit,
              "random_forest" : pipe_rf,
              "decision_tree" : pipe_dt,
              "svm" : pipe_svm,
              "xgboost" : pipe_xbgoost,
              "lgbm" : pipe_lgbm,
              "knn" : pipe_knn}

# =======================================

df_results = experimentos_ciclo_de_modelagem(dict_pipes,
                                             print_progress = False)

df_results
```

Dos modelos acima, o lightgbm e o xgboost apresentaram f1 score melhores, além disso o gap entre o f1 de treino e de teste do lgbm foi bem baixo.

Ademais, podemos notar um pouco de overfit no random forest e na decision tree. Isso pode ter acontecido principalmente em razão de não ter sido definido nenhum hiperparâmetro ao instanciar estes modelos.

Para verificar se conseguimos melhorar o random forest, vamos utilizar o random search para a otimização dos hiperparâmetros.

### Random Forest com otimização de hiperparâmetros

```python
# pipeline
pipe_rf = Pipeline([("preprocessor", preprocessor),
                    ("rf", RandomForestClassifier(n_jobs=-1, random_state=42))])

# espaço de hiperparâmetros 
# (esta definição vem do conhecimento que temos do método e de seus hiperparâmetros)
params_distributions = {"rf__n_estimators" : np.arange(100, 1501, 1), 
                        "rf__max_depth" : np.arange(2, 9, 1)}

# estratégia de cross validation 
splitter = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# montamos o objeto do random search
rand_grid_rf = RandomizedSearchCV(pipe_rf, 
                                  params_distributions, 
                                  n_iter=20,
                                  cv=splitter,
                                  scoring="f1_weighted",
                                  verbose=10,
                                  n_jobs=-1,
                                  random_state=42,
                                  return_train_score=True)

rand_grid_rf.fit(X_train, y_train)
```

```python
rand_grid_rf.best_params_
```

```python
pipe_rf = Pipeline([("preprocessor", preprocessor),
                    ("rf", RandomForestClassifier(random_state=42))]).set_params(**{'rf__n_estimators': 1417,
                                                                                    'rf__max_depth': 7})

pipe_rf.fit(X_train, y_train)

# ================================
# Passo 2

clf_metrics(pipe_rf, X_train, y_train, "treino")
print("#"*80)
clf_metrics(pipe_rf, X_test, y_test, "teste")
```

Com a otimização de hiperparâmetro conseguimos resolver o problema de overfitting e ainda trazer métricas um pouco mais próximas do xgboost

```python
df_feat_import = pd.DataFrame(pipe_rf['rf'].feature_importances_, index=preprocessor.get_feature_names_out(), columns=['feature importance'])
df_feat_import.sort_values(by="feature importance", ascending=False)
```

# Neural networks

Por fim, vamos ver se uma rede neural pode apresentar resultados tão bons.

```python
tf.random.set_seed(42)  # applied to achieve consistent results
model = Sequential(
    [
        Dense(27, activation = 'relu', input_shape=(27,),  name = "L1"),
        Dense(40, activation = 'relu',   name = "L2"),
        Dense(30, activation = 'relu',   name = "L3"),
        Dense(1, activation = 'sigmoid', name = "L4")
    ]
)
```

```python
model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(0.01),
)
```

```python
model.summary()
```

```python
early_stop = EarlyStopping(monitor = 'val_loss',
                           min_delta=0.001,
                           patience = 30,
                           verbose = 1,
                           mode = 'auto')
```

```python
model_es = KerasClassifier(model=model, verbose=1, epochs=100, validation_split = 0.3, callbacks=[early_stop])
```

```python
pipe_nn = Pipeline([("preprocessor", preprocessor),
                    ("nn", model_es)])
```

```python
pipe_nn.fit(X_train, y_train)
```

```python
# ================================
# Passo 2

clf_metrics(pipe_nn, X_train, y_train, "treino")
print("#"*80)
clf_metrics(pipe_nn, X_test, y_test, "teste")
```

Das métricas acima, verificamos que a rede neural apresenta resultados tão bons quanto o lgbm e xgboost.
